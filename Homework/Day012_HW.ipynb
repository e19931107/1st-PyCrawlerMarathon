```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, "|" ,d.find_all('em')[-1].text,"|",d.find_all('a')[-1].text)
```

```python
#2.如果想要依照類別分類，怎麼儲存會比較好？

date = [d.find(class_="date").text for d in soup.find(class_="part_list_2").find_all('h3') ]
category = [d.find_all('em')[-1].text for d in soup.find(class_="part_list_2").find_all('h3')]
title = [d.find_all('a')[-1].text for d in soup.find(class_="part_list_2").find_all('h3')]


df = pd.DataFrame({'date':date,'category':category,'title':title})
df

```

```python
# 3.哪一個類別的文章最多

list = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    i = d.find_all('em')[-1].text
    list.append(i)
#print(list)
list1 = set(list) 


print('--------')

list2 = []
list4 = []

for m in set(list1):
    print('%s 出現的次數為 %d' % (m,list.count(m)))
    list2.append(list.count(m))
    list4.append(m)

#這些頭條類別分別有幾篇po文
#print(list2)
#print('今日頭條有哪些類別：',list4)

#找出今日最多頭條類別的項目位置
print('----------')
#print(list2.index(max(list2)))
n = list2.index(max(list2))

print('哪類別的文章最多:',list4[n])
```
