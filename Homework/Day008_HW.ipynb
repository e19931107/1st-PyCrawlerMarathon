```python
import requests
# 引入函式庫
header = {
    'user-agent': 'Mozilla/5.0 (Windows Phone 10.0; Android 4.2.1; Microsoft; Lumia 950) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/46.0.2486.0 Mobile Safari/537.36 Edge/14.14263'
}
r = requests.get('https://www.dcard.tw/f',headers = header)
# 想要爬資料的目標網址
r.text
# response = r.text
# # 模擬發送請求的動作

# import json
# response = json.loads(response)
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site= "https://www.dcard.tw/f"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page)
#print(soup)

print(soup)
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site= "https://www.zhihu.com/explore"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page)
print(soup)
```
